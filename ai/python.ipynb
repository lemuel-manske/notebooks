Thinking in vectors

```python
numbers = range(1, 5)

squared = [n**2 for n in numbers]

print(squared)

# or maybe...

import numpy as np

numbers = np.array(range(1, 5))

squared = numbers ** 2

print(squared)
```

```python
numbers = np.array(range(1, 11))

by_5 = numbers * 5

mean = np.mean(by_5) # sum(by_5) / numbers.size

print(mean)
```

```python
arr = np.arange(100)

even = arr[arr % 2 == 0]

squared = even ** 2

mean = np.mean(squared)

print(mean)
```

```python
def relu(x):
    return max(0, x)

def neuron(x1, x2, w1, w2, b):
    return w1*x1 + w2*x2 + b

def hidden_layer(x, y):
    h1 = relu(neuron(x, y, 1, 0, 0))      # |
    h2 = relu(neuron(x, y, 0, 1, 0))      # _
    h3 = relu(neuron(x, y, -1, -1, 5))    # /
    return h1, h2, h3

def output_layer(h1, h2, h3):
    return h1 + h2 + h3

points = [
    (0, 0),
    (2, 2),
    (4, 4),
    (5, 0),
]

for p in points:
    h = hidden_layer(p[0], p[1])
    out = output_layer(*h)
    print(p, "->", out)
```
